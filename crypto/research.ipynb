```python
import pandas as pd

df = pd.read_csv('BTC-USD.csv')
df.describe()
```

```python
import plotly.graph_objects as go
import plotly.express as plt

fig = go.Figure()

fig.add_trace(go.Scatter(x = df['Date'], y = df['High'], name = 'High'))
fig.add_trace(go.Scatter(x = df['Date'], y = df['Low'], name = 'Low'))
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x = df['Date'], y = df['Open'], name = 'Open'))
fig.add_trace(go.Scatter(x = df['Date'], y = df['Close'], name = 'Close'))
```

```python
df['HLRange'] = abs(df['High'] - df['Low'])
df['OCRange'] = abs(df['Open'] - df['Close'])
df[['HLRange', 'OCRange']].describe()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x = df['Date'], y = df['HLRange'], name = 'HLRange'))
fig.add_trace(go.Scatter(x = df['Date'], y = df['OCRange'], name = 'OCRange'))
```

```python
ranges = df[['HLRange', 'OCRange']]
plt.scatter(df, 'HLRange', 'OCRange')
```
