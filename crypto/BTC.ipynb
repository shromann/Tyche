```python
start_date = '2021-04-21'
end_date = '2022-04-22'

df = pd.read_csv('BTC-USD.csv')

df['HLRange'] = (df['High'] - df['Low'])
df['OCRange'] = (df['Open'] - df['Close'])

df['Date'] = pd.to_datetime(df['Date'])
df = df[(df['Date'] > pd.to_datetime(start_date)) & (df['Date'] < pd.to_datetime(end_date))]

HLRange = (df['HLRange'] - df['HLRange'].mean()) / df['HLRange'].std()
OCRange = (df['OCRange'] - df['OCRange'].mean()) / df['OCRange'].std()

fig = ff.create_distplot([OCRange, HLRange], group_labels=['Open Close', 'High Low'], bin_size=[0.1, 0.1])
# fig.show()

OCRatio = df[df['OCRange'] >= df['OCRange'].mean()]['OCRange'].size / df[df['OCRange'] < df['OCRange'].mean()]['OCRange'].size
HLRatio = df[df['HLRange'] >= df['HLRange'].mean()]['HLRange'].size / df[df['HLRange'] < df['HLRange'].mean()]['HLRange'].size

OCRatio, HLRatio

for i in range(4, 12):
    start_date = f'2021-0{i}-21'
    end_date = f'2022-0{i + 1}-22'


```
